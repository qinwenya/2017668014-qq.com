```python
import mytools
import pandas as pd
```

```python
数据表 = mytools.读取SPSS数据文件(R'data\国家认同数据清理后.sav')
```

```python
# 创建年级、政治面貌和国歌升起的交互表
cross_tab = pd.crosstab(index=[数据表['年级'], 数据表['国歌升起']], 
                        columns=数据表['政治面貌'], 
                        margins=True, normalize=True)  # All 行和列的总和

cross_tab
```

```python
from scipy import stats
specific_grade = "大四"

# 筛选出特定年级的数据
df_specific_grade = 数据表[数据表['年级'] == specific_grade]

# 创建政治面貌和会隐瞒身份吗的交互表
cross_tab_specific_grade = pd.crosstab(df_specific_grade['会隐瞒身份吗'], df_specific_grade['政治面貌'])

# 进行卡方检验
chi2_grade, p_grade, dof_grade, expected_grade = stats.chi2_contingency(cross_tab_specific_grade)

print(chi2_grade,p_grade)
cross_tab_specific_grade
```
